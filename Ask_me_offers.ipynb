```python
import pandas as pd
import numpy as np
import random
from faker import Faker
```

```python
fake = Faker()
num_rows = 1000
```

```python
user_ids = random.sample(range(1, 10000), 1000)
minimum_purchase = [round(random.uniform(1000, 10000), 2) for _ in range(num_rows)]
coupons_offered = [random.randint(0, 10) for _ in range(num_rows)]
redeemed_coupons = [random.randint(0, coupons_offered[i]) for i in range(num_rows)]
user_location = [random.choice(['India','USA']) for _ in range(num_rows)]
coupon_type = [random.choice(['Percentage', 'Fixed Amount']) for _ in range(num_rows)]
coupon_usage_frequency = [random.randint(0, 10) for _ in range(num_rows)]
website_visits = [random.randint(1, 100) for _ in range(num_rows)]
time_spent_on_site = [random.randint(60, 1800) for _ in range(num_rows)]
purchase_history = [[fake.word() for _ in range(random.randint(1, 5))] for _ in range(num_rows)]
purchased_completed = [random.choice([True, False]) for _ in range(num_rows)]
```

```python
data = {
    'User ID': user_ids,
    'Minimum Purchase': minimum_purchase,
    'Coupons Offered': coupons_offered,
    'Redeemed Coupons': redeemed_coupons,
    'User Location': user_location,
    'Coupon Type': coupon_type,
    'Coupon Usage Frequency': coupon_usage_frequency,
    'Website Visits': website_visits,
    'Time Spent on Site (seconds)': time_spent_on_site,
    'Purchase History': purchase_history,
    'Purchased Completed': purchased_completed,
}
```

```python
print(data)
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv('data.csv')
```
